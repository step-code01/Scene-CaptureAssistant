```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

#files upload
from google.colab import files
uploaded_files = files.upload()

image = cv2.imread(list(uploaded_files.keys())[0])

if image is None:
  print("Error: Could not load image")
  exit()

cv2_imshow(image)
```

##Blur Check (Focus Detection)

```python
image = cv2.resize(image,(640,480))
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

variance_lpcian = cv2.Laplacian(gray,cv2.CV_64F).var()
print(f"Sharpness (Laplace variance): {variance_lpcian:.2f}")

#Interpretation example
if variance_lpcian < 100:
  print("Image may be blurry, check focus or camera shake")
else:
  print("Image appears sharp")
```
